```python
# Load the Titanic dataset
# You may have a different way to load the dataset, adjust accordingly
titanic_data = pd.read_csv('titanic.csv')
```

```python
# Display the first few rows of the dataset
print(titanic_data.head())
```

```python
# Summary statistics
print(titanic_data.describe())
```

```python
# Check for missing values
print(titanic_data.isnull().sum())
```

```python
# Distribution of passengers by class
sns.countplot(x='PClass', data=titanic_data)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Distribution of passengers by age
sns.histplot(titanic_data['Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Passenger Ages')
plt.show()
```

```python
# Survival by passenger class
sns.countplot(x='Survived', hue='PClass', data=titanic_data)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
# Survival by gender
sns.countplot(x='Survived', hue='Sex', data=titanic_data)
plt.title('Survival by Gender')
plt.show()
```

```python
sns.violinplot(x='Sex',y='Age',hue='Survived',data=titanic_data,split=True)
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='PClass',y='Age',data=titanic_data,palette='winter')
```

```python
# Plotting pairplot
sns.pairplot(titanic_data,hue='Survived',palette='bwr')
plt.show()
```
